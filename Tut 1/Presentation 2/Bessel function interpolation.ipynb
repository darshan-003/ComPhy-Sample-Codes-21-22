```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
x=np.arange(3.5,8.5,0.1)
y=sp.j1(x)
plt.plot(x,y)
plt.grid()
```

```python
def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])

    coef[:, 0] = y
    for i in range(1, n):
        for j in range(0, n-i):
            coef[j][i] = (coef[j+1][i-1]-coef[j][i-1])/(x[i+j]-x[j])

    return coef
```

```python
def newt_poly(x_data, x, coef):
    y_data = np.zeros([x_data.shape[0]])

    for i in range(0, y_data.shape[0]):
        ans = np.zeros([x.shape[0]])
        ans[0] = coef[0]
        for j in range(1, x.shape[0]):
            ans[j] = ((coef[j]/coef[j-1]) * (x_data[i]-x[j-1]) * ans[j-1])

        sum = 0
        for k in range(0, ans.shape[0]):
            sum = sum+ans[k]

        y_data[i] = sum

    return y_data
```

```python
x_vals= np.array([5,6])
y_vals = np.array([sp.j1(5), sp.j1(6)])

coef = divided_diff(x_vals, y_vals)[0, :]


x_new = np.arange(3, 8, .1)
y_new = newt_poly(x_new, x_vals, coef)


plt.figure(figsize=(12, 8))
plt.plot(x,y)
plt.plot(x_new,y_new)

ind = int(5.5-x_new[0]/0.1)

print("Interpolated Value:")
print(y_new[ind] )
plt.plot(x_new[ind],y_new[ind],'bo')
plt.plot(5.5,sp.j1(5.5),'bo')

print("Percentage Error:")
print(abs((y_new[ind]-sp.j1(5.5))/sp.j1(5.5)) *100)
```
